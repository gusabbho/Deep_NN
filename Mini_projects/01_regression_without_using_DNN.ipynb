```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([1.2, 2.5, 3.6])
y = np.array([8.5, 9.2, 10.1])
```

# definition of errors
The formula for absolute error (AE) is:

AE = |y_pred - y_true|

The formula for quadratic error (QE), also known as mean squared error (MSE), is:

QE = 1/N * Î£(y_pred - y_true)^2

# part 1: linear regression with MSE as loss function

```python
import numpy as np

def predict(w, b, X):
    return w * X + b

def mse(y, new_y):
    n = len(y)
    return 1/n * np.sum((y - new_y)**2)

def mse_derivatives(w, b, X, y, new_y):
    n = len(X)
    dE_dw = -2/n * np.sum(X * (y - new_y))
    dE_db = -2/n * np.sum(y - new_y)
    return dE_dw, dE_db

def train_linear_regression_MSE(X, y, learning_rate=0.01, num_epochs=1000):
    # Initialize the weights
    w_MSE = np.random.randn()
    b_MSE = np.random.randn()

    # Training loop
    for epoch in range(num_epochs):
        # Make a prediction
        new_y = predict(w_MSE, b_MSE, X)

        # Calculate the loss (mean squared error)
        loss = mse(y, new_y)

        # Calculate the gradients of the loss with respect to the weights
        dE_dw, dE_db = MSE_derivatives(w_MSE, b_MSE, X, y, new_y)

        # Update the weights
        w_MSE -= learning_rate * dE_dw
        b_MSE -= learning_rate * dE_db

    # Return the final weights and biases
    return w_MSE, b_MSE
```

```python
w_MSE,b_MSE = train_linear_regression_MSE(X,y,learning_rate=0.01, num_epochs=4)
```

```python
predict(w_MSE,b_MSE,X)
```

# part 2: regression model with MAE used as loss function

```python
# note that np.sign function returns the sign (+ or -) of the (actual_y - predicted_y) then we multiply that with X
def MAE_derivatives(w, b, X, y, new_y):
    N = len(X)
    dE_dw = -1/N * np.sum(np.sign(y - new_y) * X)
    dE_db = -1/N * np.sum(np.sign(y - new_y))
    return dE_dw, dE_db
```

```python
def predict(w, b, X):
    return w * X + b

def mse(y, new_y):
    N = len(y)
    return 1/N * np.sum((y - new_y)**2)

def MAE_derivatives(w, b, X, y, new_y):
    N = len(X)
    dE_dw = -2/N * np.sum(X * (y - new_y))
    dE_db = -2/N * np.sum(y - new_y)
    return dE_dw, dE_db

def train_linear_regression_MAE(X, y, learning_rate=0.01, num_epochs=4):
    # Initialize the weights
    w_MAE = np.random.randn()
    b_MAE = np.random.randn()

    # Training loop
    for epoch in range(num_epochs):
        # Make a prediction
        new_y = predict(w_MAE, b_MAE, X)

        # Calculate the loss (mean squared error)
        loss = mse(y, new_y)

        # Calculate the gradients of the loss with respect to the weights
        dE_dw, dE_db = MSE_derivatives(w_MAE, b_MAE, X, y, new_y)

        # Update the weights
        w_MAE -= learning_rate * dE_dw
        b_MAE -= learning_rate * dE_db

    # Return the final weights and biases
    return w_MAE, b_MAE
```

```python
w_MAE,b_MAE = train_linear_regression_MSE(X,y,learning_rate=0.01, num_epochs=4)
```

```python
predict(w_MAE,b_MAE,X)
```

# part 3: plotting the original data along two regression lines we trained.

```python
plt.scatter(X, y, color='red', label='data')
plt.plot(X, predict(w_MSE, b_MSE, X), color='blue', label='regression line with MSE Loss')
plt.plot(X, predict(w_MAE, b_MAE, X), color='green', label='regression line with MAE Loss')
plt.legend()
plt.show()
```
